```python
import pandas as pd
import numpy as np 
df = pd.read_csv('imdb_1000.csv')
df.head()
```

```python
 #genres in the data
df['genre'].unique()
```

```python
#average duration of each genre
df.groupby('genre')['duration'].mean()
```

```python
#average duration of each genre as pper content rating
df['content_rating'].unique()
```

```python
# filling unkown values that is nan with unknown
df['content_rating'] = df['content_rating'].fillna("Unknown")
```

```python
print("Number of content ratings - ")
print(df['content_rating'].nunique())
```

```python
print("Number of genres - ")
print(df['genre'].nunique())
```

```python
df.groupby(['genre','content_rating'])['duration'].mean()
```

```python
df.groupby(['genre','content_rating'])['duration'].mean().unstack()
```

```python
# Number of movies in each genre for each content rating.
```

```python
df.groupby(['genre'])['content_rating'].value_counts().unstack().fillna(0)
```

```python
# Maximum , minimum,mean star_rating & duration of each genre
```

```python
df.groupby('genre')[['star_rating','duration']].agg(['mean','min','max']).reset_index()
```

```python
# filter all movie infos of 'Crime' , 'action' and 'comedy'
```

```python
df[(df['genre'] == 'Crime') | (df['genre'] == 'Comedy') | (df['genre'] == 'Action')]
```

```python
#isin() => or condition for a data column 
df[df['genre'].isin(['Crime','Comedy','Action'])]
```

```python
#Filter the data from genre crime comedy action with duration more than 140 and rating less than 8.5.
#Filter all movues in which 'Tim Robbins' is an actor.
```

```python
df[df['genre'].isin(['Crime','Comedy','Action']) & ((df['duration'] >= 140) & (df['star_rating'] < 8.5))]
```
